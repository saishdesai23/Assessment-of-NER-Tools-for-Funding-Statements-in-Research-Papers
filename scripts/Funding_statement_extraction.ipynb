```python
import csv
from bs4 import BeautifulSoup as bs
```

```python
xml_path = "../data/xml/pmc_result_20210101_20210105.xml"
input_file = open(xml_path, 'r')
contents = input_file.read()
# embedding the XML with beautiful soup module

soup = bs(contents, 'xml')
```

```python
count = 0
xml_papers = soup.find_all("article")
for paper in xml_papers:
    title = paper.find_all("title")
    for t in title:
        if t.text == 'Funding':
            funding = t.find_next("p").text
            count+=1
            break
        else:
            funding = 'NA'
    print(funding)
    print("************")
```

```python
xml_papers = soup.find_all("article")
count = 0
pcount = 0
for paper in xml_papers:
    pcount+=1
    acknowledgments = paper.find("ack")
    if acknowledgments is None:
        print("NA")
    else:
        ack = acknowledgments.find("p").text
        if ack is None:
            print("NA")
        else:
            print(ack)
        acknowledgments = acknowledgments.find_next("p")
        print(acknowledgments)
        
        # Additional Funding Information from <sec> tag
        if acknowledgments.find_next().name == 'sec':
            acknowledgments = acknowledgments.find_next("sec")
            acknowledgments = acknowledgments.find_all("title")
            for ele in acknowledgments:
                if ele.text == 'Funding':
                    print(ele.text + "123")
                    print(ele.find_next("p").text)
            # funding = acknowledgments.find_next("sec").find_next("p")
        else:
            funding = "NA"  
        
        print(funding)
    
    # Funding Information from Footenote <fn> tag
    fn = paper.find_all("fn-group")
    if fn == None:
        print("NA")
    else:
        for ele in fn:
            temp = ele.find("fn", {"fn-type": "supported-by"})
            if temp == None:
                print("NA")
            else:
                print(temp)
                count+=1
    
    print("************************")
```
